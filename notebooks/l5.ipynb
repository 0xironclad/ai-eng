```python
import os
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
#  Here, we're using the responses API to get a response from the model
responses = client.responses.create(
    model="gpt-5",
    input="Hello, how are you?"
)
# print(responses.output[1].content[0].text)
print(responses.output_text)
```

```python
# Here, we're using the chat completions API to get a response from the model. This is different from the responses API we used earlier since it allows for conversation history. 
messages = [
    {"role": "system", "content": "You are a helpful coding assistant."},
    {"role": "user", "content": "Hello, how are you?"},
    {"role": "assistant", "content": "I'm doing well, thanks! How can I help you today?"},
    {"role": "user", "content": "What is recursion in programming?"},
]

completion = client.chat.completions.create(
    model="gpt-4.1",
    messages=messages
)
messages.append({"role": "assistant", "content": completion.choices[0].message.content}) # Add the assistant's response to the conversation history
print(completion.choices[0].message.content)
```
