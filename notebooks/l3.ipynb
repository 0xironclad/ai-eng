# Transforming
This notebook explores how to use Large Language Models for text transformation tasks such as language translation, spelling and grammar checking, tone adjustment, and format conversion.

````python
import os
from openai import OpenAI
import dotenv

dotenv.load_dotenv()

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
````

````python
def get_completion(prompt: str, model: str = "gpt-5.4-mini") -> str:
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content
````

````python
prompt = f"""
Translate the following English text to Swahili: 
```Hi, I would like to order a blender```
"""
response = get_completion(prompt)
print(response)
````

````python
prompt = f"""
Translate the following  text to Swahili and 
Kalenjin pirate.: 
```Come home now```
"""
response = get_completion(prompt)
print(response)
````

# Universal Translator

Here, we'll create a universal translator that can translate text to any language.

````python
# Assume we have a list of user requests in different languages
user_messages = [
  "La performance du système est plus lente que d'habitude.",  # System performance is slower than normal         
  "Mi monitor tiene píxeles que no se iluminan.",              # My monitor has pixels that are not lighting
  "Il mio mouse non funziona",                                 # My mouse is not working
  "Mój klawisz Ctrl jest zepsuty",                             # My keyboard has a broken control key
  "我的屏幕在闪烁"                                               # My screen is flashing
] 
````

````python
for issue in user_messages:
    prompt = f"""
    Tell me what language this is and 
    Translate the text to English: {issue}.
    Give the language name and the translated text in the following format:
    <language name> : <translated text>
    """
    response = get_completion(prompt)
    print(response)
````
