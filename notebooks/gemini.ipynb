# In this notebook, we'll explore the Gemini API and its capabilities.

```python
from google import genai
import os
import dotenv
dotenv.load_dotenv()

client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
```

```python
interaction = client.interactions.create(
    model="gemini-2.5-flash",
    input="Hello there! What's the weather like in Budapest?"
)
print(interaction.output_text)
```

```python
# Multi-turn conversation
chats = client.chats.create(model="gemini-2.5-flash")
# print(chats)
response = chats.send_message("Hello")
print(response.text)
response = chats.send_message("What message did i send earlier?")
print(response.text)
```
